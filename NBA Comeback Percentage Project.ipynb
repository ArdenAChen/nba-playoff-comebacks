## Data Set Column Terminology

Anything not listed here will be defined when brought up
* `Higher Seed` = team in series with the higher seed value
* `Lower seed` = team in series with the lower seed value
* `Higher Seed #` = seed value of the higher seed team
* `Lower Seed #` = seed value of the lower seed team
* `HS HCA?` = Abbreviation for Higher Seed Homecourt Advantage? Accounts for series where the higher seed does not have homecourt advantage due to a variety of reasons.
* `Round` = playoff round series took place
  * 1 = First Round
  * 2 = Second Round (Conference Semi-Finals)
  * 3 = Third Round (Conference Finals)
  * 4 = Fourth Round (Finals)
  * In some of the earlier years of the NBA, the 1st and 2nd rounds did not exist. To ensure continuity, the Finals still count as the Fourth Round, while the Conference Finals still count as the Third Round. When the 1st round does not exist, the Semi-Finals (which are the first round in that year's brackets) are still considered the Second Round.
* `Games`: games it took for the series to end
  * 7-game series can end in a range from 4 to 7 games
  * 5-game series can end in a range from 3 to 5 games
* `Comeback`: whether the team that fell down 0-2 in the series ended up winning the series or not
  * 'Y' = A comeback was made by the team down 0-2
  * 'N' = A comeback was not made by the team down 0-2
* `Champion won`: the team that won the championship that year won the series
* `Finalist won`: one of the teams that played in the finals that year won the series
* `Higher Seed Down`: whether the team that is trailing 0-2 in the series is the higher seed or not
  * 'Y' = the team trailing down 0-2 in the series is the higher seed
  * 'N' = the team trailing down 0-2 in the series is the lower seed
* `Year`: The year in which the playoff series took place
* `Winner`: the team that won the series

# Calculating the odds for a comeback based only on each different value of seed differences

This is one of the two sections about `Seed Difference` that will be used for the model. This cell only calculates the odds of a comeback given that the seed difference is one exact value. As one would expect, when the seed difference is less, there is a greater chance for a comeback scenario, though there are some outliers due to the small sample size, the trend still remains true.

```python
# Create an empty list to store the values which will be inputted into a new DataFrame
sd_value_list = []
sd_comebacks_list = []
sd_value_series_list = []
sd_cb_pctg_list = []
sd_pctg_multiplier_list = []

# For loop to calculate stats when sd_value is equal to each seed difference value
for sd_value in range(0, 8):
    # Calculate for Comeback scenario
    df_sd_value_cb = df[(df['Seed Difference'] == sd_value) & (df['Comeback'] == 'Y')]
    count_sd_value_cb = len(df_sd_value_cb)
    count_sd_value_series = len(df[df['Seed Difference'] == sd_value])

    # Calculate comeback pctg
    if count_sd_value_series != 0:
        sd_comeback_pctg = count_sd_value_cb / count_sd_value_series * 100
    else:
        sd_comeback_pctg = 0

    # Compare to baseline 7.41%
    sd_pctg_multiplier = sd_comeback_pctg / default_cb_pctg

    # Append values to lists
    sd_value_list.append(sd_value)
    sd_comebacks_list.append(count_sd_value_cb)
    sd_value_series_list.append(count_sd_value_series)
    sd_cb_pctg_list.append(round(sd_comeback_pctg, 2))
    sd_pctg_multiplier_list.append(sd_pctg_multiplier)

    # Print the Results for Comeback scenario
    print(f'When the seed difference is {sd_value}:')
    print(f'The amount of times a team has come back from a 2-0 deficit is {count_sd_value_cb}.')
    print(f'The amount of times a team has gone up 2-0 when the seed difference is {sd_value} is {count_sd_value_series}.')
    print(f'This gives teams a {round(sd_comeback_pctg, 2)}% chance to come back from down 2-0 in a series.')
    print(f'This gives teams {round(sd_pctg_multiplier, 2)} times the odds of making a comeback compared to the average 7.41% \n')

df_equals_sd_value_stats = pd.DataFrame({
    'Seed Difference': sd_value_list,
    'Comebacks': sd_comebacks_list,
    'Total Series': sd_value_series_list,
    'Comeback Pctg': sd_cb_pctg_list,
    'Percentage Multiplier': sd_pctg_multiplier_list
})

df_equals_sd_value_stats
```

Here is what each column in the DataFrame represents:
* **`Seed Difference`**: the seed difference value that is considered in the row's calculations.
* * Example: `Seed Difference == 5`: only series with a seed difference of 5 are considered in the calculations
* **`Comebacks`**: the amount of series that resulted in a comeback
* **`Total Series`**: the total amount of 0-2 series that had this seed difference value
* **`Comeback Pctg`**: the percent of the total series which were comebacks ($Comeback\;Pctg=\frac{Comebacks}{Total\;Series}$)
* **`Percentage Multiplier`**: how many times better or worse the odds of a comeback is based on the seed difference value compared to the base percentage ($7.41\%$). Calculated by dividing `Comeback Pctg` by `comeback_cb_pctg` ($Percentage\;Multiplier=\frac{Comeback\;Pctg}{\approx7.41}$). Values greater than 1 increase the odds of a comeback compared to $7.41\%$, values less than 1 decrease the odds of a comeback compared to $7.41\%$, a score of 0 means that there is no chance of a comeback, and a score of 1 means the chances are the exact same (no changes).

This DataFrame gives more context to the section above. With the seed difference range findings, the percentages were skewed because they included series where the seed difference was 1. When isolating the seed difference values, we can see that the only seed difference values that actually increase the chance of a comeback are when the seed difference is either 1 or 2. Every other value actually makes it less likely for a comeback to occur compared to the baseline. With that said though, notice that the amount of series when the Seed Difference is 0 or 6 are very low. No conclusion can be drawn on these two rows because the sample size is simply too small, especially when compared to the other rows. This still gives us 50+ series for each seed difference category in every other row, and it is intriguing how the row with the largest sample size (`Seed Difference == 1`) has the highest comeback percentage. Does this mean that the chance of a comeback should be higher for all seed difference values, or that it is truly more likely for a comeback to happen when the seed difference is 1? Moreover, it is interesting how no comebacks have occurred when the seed difference was exactly 5, despite having more total series than most other seed difference values.

### Summary

* There is not enough data to completely conclude anything about when the seed difference is 0, or 6.
* The only seed difference values that increase the odds of a comeback in comparison to the baseline are seed difference values 1 and 2.
* The odds of a comeback occurring seem to decrease if the seed difference is not 1 or 2 (though a seed difference of 0 can join the list if its sample size rises and gets more comebacks).
* There are vastly more series for when the seed difference is 1 in comparison to every other value, and it has the highest comeback percentage. This can imply one of these two possibilities:
* * The total chance of a comeback should rise for all seed differences as the sample size rises, and potentially the reason why the comeback odds are so low for the lower seed difference values are because of the smaller sample size.
  * The chance of making a comeback rises only for when the seed difference is 1 or 2, and this is backed up by a relatively large sample size.

## Calculating the coefficient of determination ($R^2$)

The coefficient of determination judges how well of a predictor the regression line is at predicting the y-values of the graph (in this case, the chance of a comeback occurring based on seed difference). A score of 0 indicates that the line of regression is not at all a good model for predicting the amount of games a series will take. A score of 1 means the line of regression perfectly predicts the values of every data point.

# Calculating the odds for a comeback when the seed difference is 0-2 compared to 3-7

The calculations themselves will look more like for the ranges `1-2` and `3-5 and 7`, but will apply to those seed difference values.

```python
# comeback odds when the seed difference is 1 or 2 (counts as range 0-2 since 0 doesn't have enough data)
count_sd_0_to_2_cb = len(df[(df['Seed Difference'].isin([1, 2])) & (df['Comeback'] == 'Y')])
count_sd_0_to_2 = len(df[df['Seed Difference'].isin([1, 2])])
sd_0_to_2_cb_pctg = count_sd_0_to_2_cb / count_sd_0_to_2 * 100
pctg_multiplier_sd_0_to_2 = sd_0_to_2_cb_pctg / default_cb_pctg

count_sd_3_to_7_cb = len(df[(df['Seed Difference'].isin([3, 4, 5, 7])) & (df['Comeback'] == 'Y')])
count_sd_3_to_7 = len(df[df['Seed Difference'].isin([3, 4, 5, 7])])
sd_3_to_7_cb_pctg = count_sd_3_to_7_cb / count_sd_3_to_7 * 100
pctg_multiplier_sd_3_to_7 = sd_3_to_7_cb_pctg / default_cb_pctg

direct_multiplier_low_vs_high_sd = sd_0_to_2_cb_pctg / sd_3_to_7_cb_pctg # directly compare to when the seed difference is range 3 to 7
direct_multiplier_high_vs_low_sd = sd_3_to_7_cb_pctg / sd_0_to_2_cb_pctg # directly compare to when the seed difference is range 0 to 2

print(f'{count_sd_0_to_2_cb} Teams have made an 0-2 comeback in the seed difference range 0 to 2.')
print(f'{count_sd_0_to_2} Teams have gone down 0-2 in this seed difference range.')
print(f'{round(sd_0_to_2_cb_pctg, 2)}% of teams have made a comeback when down 0-2 in this seed difference range.')
print(f'Teams are {round(pctg_multiplier_sd_0_to_2, 4)} times more likely to make a comeback than the default 7.41%.')
print(f'Teams are {round(direct_multiplier_low_vs_high_sd, 4)} times more likely to make a comeback when the seed difference is 0 to 2 compared to when the seed difference is 3 to 7.\n')

print(f'{count_sd_3_to_7_cb} Teams have made an 0-2 comeback in the seed difference range 3 to 7.')
print(f'{count_sd_3_to_7} Teams have gone down 0-2 in this seed difference range.')
print(f'{round(sd_3_to_7_cb_pctg, 2)}% of teams have made a comeback when down 0-2 in this seed difference range.')
print(f'Teams are {round(pctg_multiplier_sd_3_to_7, 4)} times less likely to make a comeback than the default 7.41%.')
print(f'Teams are {round(direct_multiplier_high_vs_low_sd, 4)} times less likely to make a comeback when the seed difference is 3 to 7 compared to when the seed difference is 0 to 2.')

df_pred_model_sd_range = pd.DataFrame({ # pred = predictive
    'Seed Difference Range': ['0 to 2', '3 to 7'],
    'Comebacks': [count_sd_0_to_2_cb, count_sd_3_to_7_cb],
    'Total Series': [count_sd_0_to_2, count_sd_3_to_7],
    'Comeback Pctg': [sd_0_to_2_cb_pctg, sd_3_to_7_cb_pctg],
    'Percentage Multiplier': [pctg_multiplier_sd_0_to_2, pctg_multiplier_sd_3_to_7],
    'Direct Multiplier': [direct_multiplier_low_vs_high_sd, direct_multiplier_high_vs_low_sd]
})

df_pred_model_sd_range
```

# Calculating comeback scenarios based on which playoff round it is

Here I will calculate the percent of series that were comebacks by round (the odds of a comeback in the first round, second round, third round, and fourth round).

```python
# Create a list that will be appended with the percent chance of a comeback for every round
round_value_list = []
round_cb_list = []
round_cb_pctg_list = []
round_series_list = []
round_multiplier_list = []

# For loop to calculate odds for evert round
for round_value in range(1, 5):
    # Calculate for Comeback scenario and Round
    df_round_cb = df[(df['Round'] == round_value) & (df['Comeback'] == 'Y')]
    count_round_cb = len(df_round_cb)
    count_round_series = len(df[df['Round'] == round_value])

    # Calculate comeback pctg; check for division by 0
    if count_round_series != 0:
        round_cb_pctg = count_round_cb / count_round_series * 100
    else:
        round_cb_pctg = 0

    # Calculate comeback odds multiplier
    round_multiplier = round_cb_pctg / default_cb_pctg

    # Append values to lists for DataFrame and graphs
    round_value_list.append(round_value)
    round_cb_list.append(count_round_cb)
    round_cb_pctg_list.append(round(round_cb_pctg, 2))
    round_series_list.append(count_round_series)
    round_multiplier_list.append(round_multiplier)

    # Print the Results for Comeback scenario
    print(f'When the playoff round is {round_value}:')
    print(f'The amount of times a team has come back from a 2-0 deficit is {count_round_cb}.')
    print(f'The amount of times a team has gone up 2-0 is {count_round_series}.')
    print(f'This gives teams a {round(round_cb_pctg, 2)}% chance to come back from down 0-2.\n')

# Create DataFrame
df_round_cb_stats = pd.DataFrame({
    'Round': round_value_list,
    'Comebacks': round_cb_list,
    'Total Series': round_series_list,
    'Comeback Pctg': round_cb_pctg_list,
    'Percentage Multiplier': round_multiplier_list
})

df_round_cb_stats
```

DataFrame column definitions:
* **`Round`**: The round that corresponds with the row
* **`Comebacks`**: The amount of 0-2 comebacks that have been made in that round
* **`Total Series`**: The total amount of series where a team has gone up 2-0 in that round
* **`Comeback Pctg`**: The percent of total series which were comebacks. Calculated by $Comeback\;\%=\frac{Comebacks}{Total\;Series}$
* **`Percentage Multiplier`**: how many times better or worse the odds of a comeback is in this round compared to the base percentage ($7.41\%$). Calculated by dividing `Comeback Pctg` by `comeback_cb_pctg` ($Percentage\;Multiplier=\frac{Comeback\;\%}{\approx7.41}$). Values greater than 1 increase the odds of a comeback compared to $7.41\%$, values less than 1 decrease the odds of a comeback compared to $7.41\%$, and a score of 1 means the chances are the exact same (no changes).

The odds of a comeback seem to steadily rise the further into the playoffs the series is. One thing to note though is that sample size simultaneously decreases. This might mean that the higher comeback percentages are mostly because of sample size, but it doesn't totally explain why there does seem to be an upward trend at the very least from round 1 all the way to round 3 with pretty sizeable samples. Round 4 may be an outlier but it's interesting how there's so many comebacks in so little series.

# What about when the higher seed was down in the series?

One shortcoming of seed difference is that it does not account for when the higher seed is actually down in the series. A series where the 4 seed is up 2-0 against a 1 seed would have a seed difference of 3, but obviously this would be different from a match-up where the 1 seed is up 2-0 against a 4 seed. New acronyms in the code below:
* hs = higher seed
* ls = lower seed

## When the higher seed is down

```python
# Look at cases where the higher seed was down by creating a DataFrame sorted by higher seeds being down
df_hs_down = df[df['Higher Seed Down 0-2'] == 'Y']
len_hs_down = len(df_hs_down)

# Filter for teams that have made a 2-0 comeback when the higher seed is down
df_hs_down_cb = df[(df['Higher Seed Down 0-2'] == 'Y') & (df['Comeback'] == 'Y')]
len_hs_down_cb = len(df_hs_down_cb)

# Calculate the percent of series that ended with the higher seed making a comeback
hs_down_cb_pctg = len_hs_down_cb / len_hs_down * 100

# Comeback pctg multiplier
hs_down_cb_multiplier = hs_down_cb_pctg / default_cb_pctg
```

## When the lower seed is down

```python
# Make new DataFrame that shows when the lower seed was down in a series
df_ls_down = df[df['Higher Seed Down 0-2'] == 'N']
len_ls_down = len(df_ls_down)

# Filter for cases of comebacks when the higher seed is not down
df_ls_down_cb = df[(df['Higher Seed Down 0-2'] == 'N') & (df['Comeback'] == 'Y')]
len_ls_down_cb = len(df_ls_down_cb)

# Calculate the percent of series that ended with the lower seed making a comeback
ls_down_cb_pctg = len_ls_down_cb / len_ls_down * 100

# Comeback pctg multiplier
ls_down_cb_multiplier = ls_down_cb_pctg / default_cb_pctg
```

## Calculations

```python


# Calculate the amount of times more likely a comeback is when the higher seed is down compared to lower seed, and vice versa
cb_multiplier_hs_vs_ls = hs_down_cb_pctg / ls_down_cb_pctg
cb_multiplier_ls_vs_hs = ls_down_cb_pctg / hs_down_cb_pctg

# Print the result
print(f'There were {len_hs_down} cases where the higher seed was down 0-2 in a series. Of these series, only', end = ' ')
print(f'{len_hs_down_cb} teams have made a comeback. This gives teams a {round(hs_down_cb_pctg, 2)}%')
print(f'Chance to make a comeback. Teams are {round(hs_down_cb_multiplier, 4)} times more likely to make a comeback when down 0-2 compared to the baseline 7.41%.\n')

print(f'There were {len_ls_down} cases where the lower seed was down 0-2 in a series. Of these series, only', end = ' ')
print(f'{len_ls_down_cb} teams have made a comeback. This gives a {round(ls_down_cb_pctg, 2)}% chance')
print(f'to make a comeback. Teams are {round(ls_down_cb_multiplier, 4)} times less likely to make a comeback when down 0-2 compared to the baseline 7.41%\n')

print(f'Teams are {round(cb_multiplier_hs_vs_ls, 4)} times more likely to make a comeback down 0-2 when they are the higher seed in the', end = ' ')
print(f'series compared to when they are the \nlower seed in the series.\n')
print(f'Teams are {round(cb_multiplier_ls_vs_hs, 4)} times less likely to make a comeback down 0-2 when they are the lower seed in the', end = ' ')
print(f'series compared to when they are the \nhigher seed in the series.')
```

## Accounting for Sample Size

I want to emphasize the factors that have higher sample sizes to make predictions more accurate. In order to do this, I will weigh each factor by multiplying them by a fraction that is their sample size divided by the max sample size. It will look something like:

$Pctg\;Multiplier\times(\frac{S_i}{S_m})^{0.3}$ where $S_i$ is the factor's sample size, $S_m$ is the sample size of the factor with the highest sample size, and $0.3$ is an exponent that moderates the effect to make sure estimates are not too conservative. Binary factors like `Year Range`, `Series Length`, `Higher Seed Down 0-2`, and `Seed Difference` (range) will have the same $S_m$ at 394. Factors like `Round` and `Seed Difference` (exact) will have their own $S_m$ due to the fact there are more options to choose from which inherently lower the sample size. The $S_m$ for `Round` will be $202$ (sample size for 1st round series), while the $S_m$ for `Seed Difference` will be $140$ (sample size for a `Seed Difference` of 1).

So, for example, if we have `Year Range` for 2000-2024, we will do:

$7.41\%\times1.37\times(\frac{217}{394}^{0.3})\approx8.49\%$

As another example, if we have a `Seed Difference` of 1, we will do:
$7.41\%\times1.83\times(\frac{140}{140}^{0.3})\approx13.56\%$

For the last example, since the sample size is equal to the maximum sample size, the calculation remains unchanged.

## Accounting for $R^2$ scores

There are up to 2 options affected by their $R^2$ scores:
* `Round`
* `Seed Difference` (if you choose to use exact values)

For these two, I will raise their multipliers by an exponent equal to their $R^2$ score. It will generally follow the formula:

$Comeback\;\%\times Pctg\;Multiplier^{R^2_i}$

So for example, if we have `Round` equal to 3, then we do:

$7.41\%\times1.07^{0.98}\approx7.92\%$.

## The Final Formula

After accounting for sample size and $R^2$, we arrive on this final formula:

$P_{adjusted}=P_{base}\times\prod_i(M_i\times(\frac{S_i}{S_m})^{\alpha})^{R^2_i}$

Where:
* $P_{adjusted}$ is the new expected comeback percentage,
* $P_{base}\approx7.41\%$,
* $\prod$ is the Pi-product of all factors. Think of it like the symbol $\sum$, but instead of it being a summation of all terms, it is the product of all terms. It essentially means to multiply all 5 of the percentage multipliers you choose.
* $M_i$ is the percentage multiplier of the factor,
* $i$ is just one of options of the 5 factors,
* $S_i$ is the sample size of the factor,
* $S_m$ is the maximum sample size (391 for binary factors, 202 for round, and 140 for Seed Difference (exact)),
* $\alpha=0.3$,
* $R^2_i$ is the $R^2$ score of the factor (only applies to `Round` and `Seed Difference` (exact)).

## Estimating how many more comebacks there would have been if there were more playoff series in the early years

It is probable that there are substantially less comebacks in earlier years because there are simply less opportunities. I want to estimate how many more potential comebacks there could have been before 1984, which is the cut-off point in roughly having the modern playoff bracket.

### Eligible Series

When I refer to eligible series, I just mean **the total amount of 5-game and 7-game series in any given year.** So for example, 2024 had eight 1st-round series, four 2nd-round series, two 3rd-round series, and one Finals series, each being 7-game series, adding up to 15 "eligible series". Another example is the 1980 playoffs. In 1980, the first round was only a 3-game series, plus there were only four 1st-round series, so we exclude the entire first round. This gives 7 "eligible series". Years 1984 and after all have 15 eligible series, with prior years having 7 or less eligible series.

### How can we account for this?

The goal is to estimate how many more comebacks there could have been if there were more eligible series each year. To account for the difference in eligible series, I'm going to add a multiplier to the comebacks made in a few select years. There are 15 total 7-game series in the modern NBA format. For some years where there are less eligible series and comebacks, I will be multiplying the amount of comebacks made in that year by 15 divided by the amount of 7-game or 5-game series in that year. It would look something like this:
$$ Comebacks\;Made\times\frac{15}{Total\;Eligible\;Series}  $$
So, coming back to 1980, only 7 of the series in that year's playoffs were eligible for an 0-2 comeback, so I would multiply:
$$ 0\%\times\frac{15}{8}=0\% $$
Of course 0 times any number would stay 0, but this might matter for other years. I will store this new number in a new DataFrame column called `Adjusted Comeback Multiplier`, which will be multiplied by the amount of comebacks made that year and give a new number stored in the column `Adjusted Estimated Comebacks`. This new column is the expected amount of comebacks given there were more series that year (though I will not be treating it as so for a few reasons I will explain later). Another column will be made called `Estimated Extra Comebacks` which will show the expected amount of extra comebacks that would have been made.

### Years affected:

* 1956: 3 eliible series
* 1969: 7 eligible series
* 1971: 7 eligible series
* 1977: 7 eligible series

### 1956

There was a comeback made in 1956, but using the method I outlined above, the new comeback number estimate is outlandish. Here's the math:
$$ 1\times\frac{15}{3}$$
$$ =1\times5=5 $$
This method makes its calculations off the assumption that every other series would go just like the one where the comeback occurred. However, that simply cannot be assumed to be the case. Rather, this method would better be utilized to see how many more comebacks could have been made in the entire span of 1947-1983. I do not think there would have been 5 comebacks in 1956 if there were more series, but if every year from 1947-1983 had 15 series, there may have been 5 more comebacks in that timespan just based on that figure by itself.

### Flaws With this Method

Naturally, this is not a perfect way to account for different play-off formats. Years in the modern era are going to have an inherent advantage since they have the most series. Also, with this small of a sample from year to year, it's a little dishonest to say more comebacks would have happened in that year specifically at the same rate if there were 15 eligible series, and this definitely overcompensates for the early years since there are so little "eligible series". ultimately, the important thing to remember is that comebacks were sometimes rarer in the older eras because there were simply less series where a comeback would have been possible, so it's much more likely for any given year to not have a comeback at all. This also works the other way, as 1956 has a comeback percentage of 100%, though there was only one time in that year's playoffs where a team went down 0-2.

### Potential Improvements

I did not exclude 5-game series, and later on we will see how 5-game series do have an effect on the odds of a team coming back from a 2-0 deficit.

### Results displayed in a DataFrame that adjusts for "eligible series"

```python
# Create a new DataFrame column that stores the values which will be multiplied by 0-2 series comebacks to get estimated extra comebacks
df_year_freqs['Adjusted Comeback Multiplier'] = 1.0 # Default value for every year will be 1 unless there is a comeback in a year with less than 15 eligible series

# Use .loc to change rows 1956, 1969, 1971, and 1977's multipliers
df_year_freqs.loc[df_year_freqs['Year'] == 1956, 'Adjusted Comeback Multiplier'] = 15/3
df_year_freqs.loc[df_year_freqs['Year'].isin([1969, 1971, 1977]), 'Adjusted Comeback Multiplier'] = 15/7

# Calculate adjusted estimated comebacks by multiplying 0-2 Series Comebacks by adjusted comeback multiplier
df_year_freqs['Adjusted Estimated Comebacks'] = df_year_freqs['0-2 Series Comebacks'] * df_year_freqs['Adjusted Comeback Multiplier']

# Calculate estimated extra comebacks by subtracting adjusted estimated comebacks by 0-2 series comebacks
df_year_freqs['Estimated Extra Comebacks'] = df_year_freqs['Adjusted Estimated Comebacks'] - df_year_freqs['0-2 Series Comebacks']

# Only display the changed rows
adjusted_rows = df_year_freqs[df_year_freqs['Year'].isin([1956, 1969, 1971, 1977])]
adjusted_rows.index = (1, 2, 3, 4)
adjusted_rows
```

```python
# Adding up each year's estimated extra comebacks to see how many extra comebacks we could have expected if there were more eligible series
sum_estimated_extra_comebacks = adjusted_rows['Estimated Extra Comebacks'].sum()
print(f'The amount of extra comebacks that we could have expected if there were more eligible series is: {sum_estimated_extra_comebacks}')
```

Considering that 1969 only had 2 comebacks, and 1977, 1971, and 1956 had only 1 each, we could've expected something more like **8 to 9 more comebacks** in total for the years before 1984, which was the first year to have the standard 16 teams and 15 eligible series that were 5-game or 7-game series. This is not an unequivocal proof that playoff structuring is the reasons that so few comebacks happened in the past, but it shows that it definitely played a factor, and this metric definitely was not perfect, only a rough estimation since we can't do anything about years where 0 comebacks were made in the first place. **My guess is that there could have been up to 9 more comebacks from 1947-1983 if every year's playoffs was structured the same way it has roughly been since 1983**

## Which franchise was the best at making comebacks

Let's see which teams had the highest winning percentage when down 0-2.

```python
# For the first three hashtagged segments: I'm basically adding a new column to our df which lists the name of the team that is down 0-2 in every series
# Case 1: Team is the higher seed and the higher seed is down 0-2
df_display_hs_down = df[df['Higher Seed Down 0-2'] == 'Y']
df_display_hs_down['Team'] = df_display_hs_down['Higher Seed']

# Case 2: Team is the lower seed and the higher seed is NOT down 0-2
df_display_ls_down = df[df['Higher Seed Down 0-2'] == 'N']
df_display_ls_down['Team'] = df_display_ls_down['Lower Seed']

# Combine both cases (resulting df is the same, just in 'Team' column it now shows which team is trailing 0-2 in the series)
df_display_trailing_team = pd.concat([df_display_hs_down, df_display_ls_down])

# Calculate total comeback chances for each team
df_count_cb_chances = df_display_trailing_team['Team'].value_counts().reset_index()
df_count_cb_chances.columns = ['Team', 'Total Comeback Chances'] # Creates new DataFrame that shows total comeback opportunities for every team

# Calculate successful comebacks for each team
df_count_successful_cb = df_display_trailing_team[df_display_trailing_team['Comeback'] == 'Y']['Team'].value_counts().reset_index()
df_count_successful_cb.columns = ['Team', 'Successful Comebacks'] # Creates new DataFrame that shows total comebacks for every team

# Merge the total chances and successful comebacks DataFrames
df_franchise_cb_stats = pd.merge(df_count_cb_chances, df_count_successful_cb, on = 'Team', how = 'left')
df_franchise_cb_stats['Successful Comebacks'] = df_franchise_cb_stats['Successful Comebacks'].fillna(0).astype(int)

# Map current franchises to their former teams so then I can add them all together
franchise_mapping = {
    'ATL + former': ['ATL', 'STL (ATL)'],
    'BRK + former': ['BRK', 'NJN (BKN)'],
    'DET + former': ['DET', 'FTW (DET)'],
    'GSW + former': ['GSW', 'PHW (GSW)', 'SFW (GSW)'],
    'HOU + former': ['HOU', 'SDR (HOU)'],
    'LAC + former': ['LAC', 'BUF (LAC)'],
    'LAL + former': ['LAL', 'MNL (LAL)'],
    'NOP + former': ['NOP', 'NOH (NOP)', 'CHH (NOP)'],
    'OKC + former': ['OKC', 'SEA (OKC)'],
    'PHI + former': ['PHI', 'SYR (PHI)'],
    'SAC + former': ['SAC', 'KCK (SAC)', 'CIN (SAC)'],
    'WAS + former': ['WAS', 'BAL (WAS)', 'WSC (WAS)']
}

# Calculate total chances and comebacks for each franchise + former
franchise_stats = []

for current_name, former_names in franchise_mapping.items():
    franchise_total_chances = 0
    franchise_successful_comebacks = 0

    for name in former_names:
        if name in df_franchise_cb_stats['Team'].values:
            franchise_total_chances += df_franchise_cb_stats[df_franchise_cb_stats['Team'] == name]['Total Comeback Chances'].values[0]
            franchise_successful_comebacks += df_franchise_cb_stats[df_franchise_cb_stats['Team'] == name]['Successful Comebacks'].values[0]

    franchise_stats.append({
        'Team': current_name,
        'Total Comeback Chances': franchise_total_chances,
        'Successful Comebacks': franchise_successful_comebacks,
        'Comeback Pctg': (franchise_successful_comebacks / franchise_total_chances) * 100 if franchise_total_chances > 0 else 0
    })

# Convert franchise_stats to df, append to df_franchise_cb_stats
franchise_stats_df = pd.DataFrame(franchise_stats)
df_franchise_cb_stats = pd.concat([df_franchise_cb_stats, franchise_stats_df], ignore_index = True)

# Calculate the comeback percentage
df_franchise_cb_stats['Comeback Pctg'] = (df_franchise_cb_stats['Successful Comebacks'] / df_franchise_cb_stats['Total Comeback Chances']) * 100

# Sort the DataFrame by 'Comeback Pctg' in descending order
df_franchise_cb_stats = df_franchise_cb_stats.sort_values(by = 'Comeback Pctg', ascending = False)

# Change index for DataFrame to go from 1 to 60
index_df_franchise_cb_stats = range(1, 1 + len(df_franchise_cb_stats))
df_franchise_cb_stats.index = index_df_franchise_cb_stats

df_franchise_cb_stats
```

### "[team name] + former" Rows
Rows that have a team name then ' + former' after, like "`GSW + former`" for example. **This row takes all the former names for the current franchise and adds them all up**, so all the values for `GSW`, `PHW (GSW)` and `SFW (GSW)` will be added up in "`GSW + former`".

Columns:
* `Team`: the franchise's name
* `Total Comeback Chances`: the total amount of times a team has fallen down 0-2. For this DataFrame, the way I had to calculate this number was odd but I'll explain it further in the code (Case 1 and Case 2 if you are interested)
* `Successful Comebacks`: the total amount of comebacks a team has after falling down 0-2.
* `Comeback Pctg`: the percent chance a team will make a comeback, given that they are already down 0-2 in a series. Calculated by dividing 'Successful Comebacks' by 'Total Comeback Chances'

Of all the teams, the Clippers and the Fort Wayne Pistons (before being renamed) have the highest comeback percentage out of all teams at $33.33\%$. ## Which teams were most and least prone to allowing comebacks?

Below we will look at teams that have blown the most 2-0 leads.

```python
# Case 1: Team is the lower seed and the higher seed is down 0-2 (now recording the team going up 2-0 instead of the team down 0-2)
df_display_ls_up = df[df['Higher Seed Down 0-2'] == 'Y']
df_display_ls_up['Team'] = df_display_ls_up['Lower Seed']

# Case 2: Team is the higher seed and the higher seed is NOT down 0-2 (now recording the team going up 2-0 instead of the team down 0-2)
df_display_hs_up = df[df['Higher Seed Down 0-2'] == 'N']
df_display_hs_up['Team'] = df_display_hs_up['Higher Seed']

# Combine both cases (resulting df now displays the name of the team up 2-0 in 'Team' category)
df_display_leading_team = pd.concat([df_display_ls_up, df_display_hs_up])

# Count number of 2-0 series leads for each team
df_count_series_leads = df_display_leading_team['Team'].value_counts().reset_index()
df_count_series_leads.columns = ['Team', '2-0 Series Leads'] # Creates new DataFrame that shows amount of 2-0 series leads for every team

# Calculate number of successful comebacks made against every team
df_count_blown_series = df_display_leading_team[df_display_leading_team['Comeback'] == 'Y']['Team'].value_counts().reset_index()
df_count_blown_series.columns = ['Team', 'Blown Series Count'] # Creates new DataFrame that shows amount of 2-0 series that were blown by each team

# Merge the 2-0 Series Leads and Series Blown DataFrames
df_blown_series_stats = pd.merge(df_count_series_leads, df_count_blown_series, on = 'Team', how = 'left')
df_blown_series_stats['Blown Series Count'] = df_blown_series_stats['Blown Series Count'].fillna(0).astype(int)

# Map current franchises to their former teams so then I can add them all together
franchise_mapping_blown_series = {
    'ATL + former': ['ATL', 'STL (ATL)'],
    'BRK + former': ['BRK', 'NJN (BKN)'],
    'DET + former': ['DET', 'FTW (DET)'],
    'GSW + former': ['GSW', 'PHW (GSW)', 'SFW (GSW)'],
    'LAL + former': ['LAL', 'MNL (LAL)'],
    'NOP + former': ['NOP', 'NOH (NOP)', 'CHH (NOP)'],
    'OKC + former': ['OKC', 'SEA (OKC)'],
    'PHI + former': ['PHI', 'SYR (PHI)'],
    'SAC + former': ['SAC', 'ROC (SAC)'],
}

# Calculate total chances and comebacks for each franchise + former
franchise_blown_lead_stats = []

for current_name, former_names in franchise_mapping_blown_series.items():
    franchise_series_leads = 0
    franchise_blown_series = 0

    for name in former_names:
        if name in df_blown_series_stats['Team'].values:
            franchise_series_leads += df_blown_series_stats[df_blown_series_stats['Team'] == name]['2-0 Series Leads'].values[0]
            franchise_blown_series += df_blown_series_stats[df_blown_series_stats['Team'] == name]['Blown Series Count'].values[0]

    franchise_blown_lead_stats.append({
        'Team': current_name,
        '2-0 Series Leads': franchise_series_leads,
        'Blown Series Count': franchise_blown_series,
        'Blown Series Pctg': (franchise_blown_series / franchise_series_leads) * 100 if franchise_series_leads > 0 else 0
    })

# Convert franchise_blown_lead_stats to df, append to df_blown_series_stats
df_franchise_blown_lead_stats = pd.DataFrame(franchise_blown_lead_stats)
df_blown_series_stats = pd.concat([df_blown_series_stats, df_franchise_blown_lead_stats], ignore_index = True)

# Calculate the Blown Series Pctg (% of series won by the opposing team after 'Team' leads 2-0, where 'Team' blows a 2-0 series lead)
df_blown_series_stats['Blown Series Pctg'] = (df_blown_series_stats['Blown Series Count'] / df_blown_series_stats['2-0 Series Leads']) * 100

# Calculate Series Clinch Percentage (% of series won by 'Team' after leading 2-0)
df_blown_series_stats['Clinch Series Pctg'] = 100 - df_blown_series_stats['Blown Series Pctg']

# Sort the DataFrame by 'Blown Series Pctg' in descending order (instead of 'Blown Series Pctg', you can sort by any other column name)
df_blown_series_stats = df_blown_series_stats.sort_values(by = 'Blown Series Pctg', ascending = False)

# Change index for DataFrame to go from 1 to 60
index_df_blown_series_stats = range(1, 1 + len(df_blown_series_stats))
df_blown_series_stats.index = index_df_blown_series_stats

df_blown_series_stats
```

This DataFrame is very similar to the previous DataFrame, except now this DataFrame records which teams allowed the most comebacks after going up 2-0 in a series.
* `Team`: Name of the franchise abbreviated
* `2-0 Series Leads`: The number of times the team has gone up 2-0 in a series
* `Blown Series Count`: The number of times the team has lost after going up 2-0 in a series
* `Blown Series Pctg`: The percent of 2-0 series that were lost after going up 2-0 in a series
* `Clinch Series Pctg`: opposite of Blown Series Pctg, the percent of 2-0 series that were won after going up 2-0 in a series

The Lakers and Celtics unsurprisingly have the most amount of series where they have gone up 2-0. Both also have low `Blown Series Pctg`, since they each have only blown 2 series when up 2-0 despite the increase in opportunities to do so. This is unsurprising since these two franchises were dominant in eras where comebacks were less common. Of the teams with 20+ series with a 2-0 lead, the Knicks have the highest Blown Series Pctg at 15%, as 3 of their 20 series were blown (most recently Knicks vs Pacers in 2024). The Jazz have blown an equal amount of 2-0 series leads at 3, but with only 15 qualifying series, giving them a Blown Series Pctg of 20%. Percentages tend to get out of hand as the "2-0 Series Lead" values get lower, mainly due to the fact that the teams at the lower end have a smaller sample size.

If we are looking at total 2-0 series leads blown, there is one team that stands out: the Phoenix Suns. In 14 series where they have led 2-0 in the series, 4 of the series were blown, indicating that Phoenix blows the most 2-0 series leads, which is highly unlucky for Suns fans. Most of the other teams are in the range of 0 to 2 leads blown when up 2-0, with NYK and UTA at 3 leads blown being the only other teams not in that range.

Observing Blown Series Pctg, the Clippers have the highest percentage at 66.67%, mainly due to blowing 2 leads in 3 series after leading 2-0.
